```python
quartermile_vacant_bldg_counts = pd.read_csv('quartermile_vacant_bldg_counts.csv')
quartermile_vacant_bldg_counts.index = quartermile_vacant_bldg_counts['School']
quartermile_vacant_bldg_counts.head()
```

```python
quartermile_vacant_parcels = pd.read_csv('quartermile_intersect_UTM40_attributes.csv')
quartermile_vacant_parcels.head()
```

```python
quartermile_vacant_sums = quartermile_vacant_parcels.groupby('School').sum()
quartermile_vacant_sums['bldg_counts'] = quartermile_vacant_bldg_counts['PNTCNT']
quartermile_vacant_sums
```

```python
quartermile_vacant_parcels.groupby('School').sum()['area'].hist(bins=50)
```

```python
import plotly.plotly as py

title = 'Baltimore City Schools Surrounded by Blight'

trace1 = Scatter(x=quartermile_vacant_sums['bldg_counts'],
                 y=quartermile_vacant_sums['area'],
                 text = quartermile_vacant_sums.index,
                 mode = 'markers',
                 marker = Marker(size = 15,
                                 sizemode = 'diameter',
                                 #sizeref = schoolage_pop.max()/1000,
                                 opacity = 0.4,
                                 color = 'red')
                 )

layout = Layout(
    title = title,
    xaxis = XAxis(range = [0,800],
                  title = 'Number of Vacant Buildings in 1/4 Mile Radius',
                  showgrid = False,
                  showline = False,
                  zeroline = False,
                  autotick = True,
                  showticklabels = True),
    yaxis = YAxis(range = [0,125000],
                  title = 'Area (m^2) of Vacant Parcels in 1/4 Mile Radius',
                  showgrid = False,
                  showline = False,
                  zeroline = False,
                  autotick = True,
                  showticklabels = True),
    autosize = False,
    width=500,
    height=500,
    font = Font(size=14),
    hovermode = 'closest',
    showlegend=False,
    margin=Margin(
        l=75,
        r=75,
        b=50,
        t=50,
        pad=0)
)

data = Data([trace1])
fig = Figure(data=data, layout=layout)
py.iplot(fig, filename=title)

```
